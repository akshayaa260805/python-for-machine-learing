```python
# Importing Modules
from sklearn import datasets
import matplotlib.pyplot as plt

# Loading dataset
iris_data = datasets.load_iris()

# Available methods on dataset
print("Methods:\n",dir(iris_data))

# Features
print("\nFeatures:\n",iris_data.feature_names)

# Targets
print("\nTargets:\n",iris_data.target)

# Target Names
print("\nTarget names:\n",iris_data.target_names)

# Dataset Slicing
x_axis = iris_data.data[:, 0] # Sepal Length
y_axis = iris_data.data[:, 2] # Petal Length

# Plotting
plt.scatter(x_axis, y_axis, c=iris_data.target)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Iris Dataset: Sepal Length vs Petal Length')
plt.show()
```
